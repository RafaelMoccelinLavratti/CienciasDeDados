```python
import matplotlib.pyplot as plt
anos = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 544.3, 1075.9, 2862.5, 5979.6, 10289.7, 14580.3]
plt.plot(anos, gdp, color='red', marker='o', linestyle='solid')
plt.title("GDP Nominal")
plt.ylabel("Bilhoes de R$")
plt.show()
```

```python
import matplotlib.pyplot as plt
filmes = ["Vingadores" , "Gente Grande", "Casablanca" , "Cada de Papel" , "Azar"]
num_oscar  = [5,11,3,8,10]
xs = [i + 0.2 for i, _ in enumerate(filmes)]
plt.bar(xs, num_oscar)
plt.ylabel('# de Premiações')
plt.title("Meus filmes favoritos")
plt.xticks([i + 0.2 for i, _ in enumerate(filmes)], filmes)
plt.show() 
```
